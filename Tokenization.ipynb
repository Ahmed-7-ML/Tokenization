## 2. Build Tokenizer

special_tokens:

- [UNK] → Unknown word
- [CLS] → Sentence beginning (important in classification)
- [SEP] → Sentence separator
- [PAD] → Padding
- [MASK] → For masked language modeling

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE   # Byte-Pair Encoding
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

# 1. Initializing the Tokenizer with the BPE Model (Use [UNK] for any unknown tokens)
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))

# 2. Determine how to initially divide the sentence (by spaces)
tokenizer.pre_tokenizer = Whitespace()

# 3. Setting up the trainer with special symbols
trainer = BpeTrainer(
    vocab_size=5000,    # Vocabulary Size = 5000 Tokens
    show_progress=True,
    special_tokens=["[UNK]", "[CLS]", "[SEP]", '[PAD]', '[MASK]']
)

# 4. Training on the medical record
# He reads all the medical sentences.
# He begins to learn: Most frequently occurring letters/syllables
# He gradually integrates them (BPE)

files = ["medical_data.txt"]
tokenizer.train(files, trainer)

# 5. Save the Tokenizer for later use -> You can use it after that without further training.
tokenizer.save("medical_tokenizer.json")
```

```python
# type_ids
# 🔹 Used when you have two statements

# For example:
# [CLS] Statement 1 [SEP] Statement 2 [SEP]

# It will be:
# [0, 0, 0, 1, 1]
# 0 → Statement 1
# 1 → Statement 2

output.type_ids
```

```python
# It tells you where each token came from in the original text.

# For example:
# [(0, 6), (7, 14), (15, 26)]

# Meaning:
# Token 1 from the character 0 → 6
# Token 2 from 7 → 14
# Token 3 from 14 → 26

# 💡 Useful for:
# NER (Entity Extraction)
# Highlighting Text

output.offsets
```

```python
# It explains the special tokens.

# For example:
# [0, 0, 0, 1, 1]
# 0 → Regular Token
# 1 → Special token such as [CLS] or [SEP]

output.special_tokens_mask
```

## 4. Practical value (compared to GPT)
- Compare your tokenizer with GPT's tokenizer (which uses the tiktoken library).

```python
import tiktoken

test_text = "تصلب الشرايين التاجية"

# 1. Hashing using your custom tokenizer
custom_tokens = tokenizer.encode(test_text).tokens
print(f"Custom Tokenizer ({len(custom_tokens)} tokens): {custom_tokens}")

# 2. Hashing using GPT-4 Tokenizer
enc = tiktoken.encoding_for_model("gpt-4")
gpt_tokens = enc.encode(test_text)
# IDs
print(f"GPT-4 Tokenizer ({len(gpt_tokens)} tokens): {gpt_tokens}")
```

```python
# Tokens
tokens = [enc.decode([t]) for t in gpt_tokens]
print(tokens)
```

```python
# Raw Bytes
tokens_bytes = [enc.decode_single_token_bytes(t) for t in gpt_tokens]
print(tokens_bytes)
```

💡 Why did your tokenizer do this?

Probably because:
1. Low data size
2. Relatively small vocabulary size
3. Insufficient repetition of medical terms
4. BPE hasn't yet taught these words as complete units

🚀 How to improve it?

1️⃣ Increase the data (thousands of medical sentences, not just hundreds)

2️⃣ Increase the vocabulary size (vocab_size=10000 or 20000)

3️⃣ Repeat important terms

4️⃣ Use Unigram instead of BPE (sometimes better for Arabic)
